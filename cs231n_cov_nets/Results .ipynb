```python
import pandas as pd
import numpy as np

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rows = pd.read_csv("results.txt")
```

```python
rows.head()
```

```python
pd.crosstab([rows.grpIndex, rows.algo], [rows.params], aggfunc=np.max, 
            values=rows.valAcc, margins=False).plot(kind='bar', sort_columns=True, rot=0, 
         figsize = (15,10), grid=True, title="Validation accuracy for various algorithms").legend(
    loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
pd.crosstab([rows.grpIndex,rows.algo], rows.params, aggfunc=np.max ,
            values=rows.valAcc, margins=False).plot(kind='bar', sort_columns=True,
         figsize = (20,20), grid=True).legend(
    loc='center left', bbox_to_anchor=(1, 0.5))
```
