```python
import pandas as pd
```

```python
history = pd.read_json("history.json")
```

```python
history
```

```python
titles = history[history["title"] != '']["title"]
```

```python
#titles.encode('utf-8').strip()
x = titles.values
y = [t.encode('utf-8').strip() for t in x]

```

```python
f = open('input.txt', 'w')
```

```python
for title in y:
    f.write(title)
    f.write("\n")
f.close()
```
